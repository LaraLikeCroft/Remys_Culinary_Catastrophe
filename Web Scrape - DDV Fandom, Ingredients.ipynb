```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://dreamlightvalleywiki.com/Ingredients'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
#soup.find('table')
```

```python
find_table = soup.find_all('table')[0]
#print(find_table)
```

```python
find_ingredients_table = soup.find('table', class_ = 'wikitable sortable')
#print(find_ingredients_table)
```

```python
table = soup.find_all('table')[0]
#print(table)
```

```python
ingredients = table.find_all('th')
#print(ingredients)
```

```python
ingredients_headers = [headers.text.strip() for headers in ingredients]
#print(ingredients_headers)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = ingredients_headers)
df
```

```python
column_data = table.find_all('tr')
#print(column_data)
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    if individual_row_data[0] != "":
           continue
    
    
    length = len(df)
    df.loc[length] = individual_row_data
```
